```python
from PIL import Image

path = ('/content/(131)IMG_1.jpg')
im = Image.open(path)

pix = np.array(im)
```

```python
pix.shape
```

```python
plt.imshow(im)
plt.show()

#이미지 흑백으로 열기
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)

plt.imshow(image_bw,'gray')
plt.show()


#이미지 red, blue로 열기
plt.imshow(image_bw,'RdBu')
plt.show()

#이미지 jet로 열기
plt.imshow(image_bw,'jet')
plt.show()

#color bar 추가하기
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.show()
```
